# BigQuery Timestamp Fetch (pandas_gbq)
Reads `data/input/seeds.csv` and retrieves block timestamps via pandas-gbq.

```python
import pandas as pd
import pandas_gbq

seeds = pd.read_csv('../data/input/seeds.csv', header=None, names=['txid'])
tx_str = ','.join(f'"{t}"' for t in seeds['txid'].tolist())
query = f"""
SELECT TO_HEX(`hash`) AS txid, block_timestamp
FROM `bigquery-public-data.crypto_bitcoin.transactions`
WHERE TO_HEX(`hash`) IN ({tx_str})
"""
project_id = pandas_gbq.context.project
timestamps = pandas_gbq.read_gbq(query, project_id=project_id)
timestamps.head()
```
