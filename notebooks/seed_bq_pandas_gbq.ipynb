# Query with pandas_gbq

Uses `pandas_gbq.read_gbq` to fetch block timestamps for txids listed in `data/input/seeds.csv`.

```python
from pathlib import Path
import pandas as pd
import pandas_gbq

BASE = Path.cwd().parent
INPUT_DIR = BASE / 'data' / 'input'
df = pd.read_csv(INPUT_DIR / 'seeds.csv')
txids = df['txid'].dropna().tolist()
txids[:5]
```

```python
txids_str = ','.join(f"'{{t}}'" for t in txids)
query = f'''
SELECT TO_HEX(`hash`) AS txid, block_timestamp
FROM `bigquery-public-data.crypto_bitcoin.transactions`
WHERE TO_HEX(`hash`) IN ({txids_str})
'''
result = pandas_gbq.read_gbq(query)
df = df.merge(result, on='txid', how='left')
df.head()
```
