# Use BigQuery Client with array parameter to fetch block timestamps


```python
import pandas as pd
from google.cloud import bigquery

seeds = pd.read_csv('../data/input/seeds.csv')
client = bigquery.Client()
query = '''
SELECT t.hash AS txid, b.timestamp AS block_timestamp
FROM `bigquery-public-data.crypto_bitcoin.transactions` AS t
JOIN `bigquery-public-data.crypto_bitcoin.blocks` AS b
ON t.block_id = b.number
WHERE t.hash IN UNNEST(@txids)
'''
job_config = bigquery.QueryJobConfig(
    query_parameters=[bigquery.ArrayQueryParameter('txids', 'STRING', seeds['txid'].tolist())]
)
result_df = client.query(query, job_config=job_config).to_dataframe()
result_df.head()
```
