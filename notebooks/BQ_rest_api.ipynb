# Use BigQuery REST API to fetch block timestamps


```python
import pandas as pd
import google.auth
import google.auth.transport.requests
import requests

seeds = pd.read_csv('../data/input/seeds.csv')
credentials, project = google.auth.default(scopes=['https://www.googleapis.com/auth/bigquery'])
credentials.refresh(google.auth.transport.requests.Request())
headers = {'Authorization': f'Bearer {credentials.token}', 'Content-Type': 'application/json'}
query = '''
SELECT t.hash AS txid, b.timestamp AS block_timestamp
FROM `bigquery-public-data.crypto_bitcoin.transactions` AS t
JOIN `bigquery-public-data.crypto_bitcoin.blocks` AS b
ON t.block_id = b.number
WHERE t.hash IN UNNEST(@txids)
'''
body = {
    'query': query,
    'useLegacySql': False,
    'parameterMode': 'NAMED',
    'queryParameters': [{
        'name': 'txids',
        'parameterType': {'type': 'ARRAY', 'arrayType': {'type': 'STRING'}},
        'parameterValue': {'arrayValues': [{'value': tx} for tx in seeds['txid'].tolist()]}
    }]
}
response = requests.post(
        f'https://bigquery.googleapis.com/bigquery/v2/projects/{project}/queries',
        headers=headers, json=body
    )
rows = response.json().get('rows', [])
result_df = pd.DataFrame([{'txid': r['f'][0]['v'], 'block_timestamp': r['f'][1]['v']} for r in rows])
result_df.head()
```
