# BigQuery magic example

Demonstrates the `%%bigquery` Jupyter cell magic to fetch block timestamps for txids in `data/input/seeds.csv`.

```python
%load_ext google.cloud.bigquery
from pathlib import Path
import pandas as pd

BASE = Path.cwd().parent
INPUT_DIR = BASE / 'data' / 'input'
df = pd.read_csv(INPUT_DIR / 'seeds.csv')
txids = df['txid'].dropna().tolist()
```

```python
%%bigquery result_df --params {"txid_list": txids}
SELECT TO_HEX(`hash`) AS txid, block_timestamp
FROM `bigquery-public-data.crypto_bitcoin.transactions`
WHERE TO_HEX(`hash`) IN UNNEST(@txid_list)
```

```python
df = df.merge(result_df, on='txid', how='left')
df.head()
```
