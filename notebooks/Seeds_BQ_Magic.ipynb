# Query seeds.csv using BigQuery magic

Uses the %%bigquery IPython magic.

```python

from pathlib import Path
import pandas as pd
from google.cloud import bigquery
from IPython import get_ipython

base = Path('..').resolve()
seeds = pd.read_csv(base / 'data' / 'input' / 'seeds.csv')
txids = seeds['txid'].dropna().tolist()

ipython = get_ipython()
ipython.run_line_magic('load_ext', 'google.cloud.bigquery')
# parameters are passed as JSON string
params = {'txid_list': txids}
```

```python

%%bigquery result --params {"txid_list": txids}
SELECT TO_HEX(`hash`) AS txid, block_timestamp
FROM `bigquery-public-data.crypto_bitcoin.transactions`
WHERE TO_HEX(`hash`) IN UNNEST(@txid_list)
```
