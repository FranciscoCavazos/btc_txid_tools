# Query seeds.csv with BigQuery client

Uses google-cloud-bigquery client and helper util.

```python

from pathlib import Path
import pandas as pd
from bq_utils import get_bigquery_client, fetch_tx_timestamps

base = Path('..').resolve()
seeds = pd.read_csv(base / 'data' / 'input' / 'seeds.csv')
txids = seeds['txid'].dropna().tolist()

client = get_bigquery_client()
timestamps = fetch_tx_timestamps(client, txids)
pd.DataFrame(list(timestamps.items()), columns=['txid', 'block_timestamp'])
```
