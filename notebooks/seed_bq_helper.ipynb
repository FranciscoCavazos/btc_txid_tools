# Fetch block timestamps with helper

Reads `data/input/seeds.csv` and uses `bq_utils.fetch_tx_timestamps` to get block timestamps from the BigQuery public Bitcoin dataset.

```python
from pathlib import Path
import sys
import pandas as pd

BASE = Path.cwd().parent
INPUT_DIR = BASE / 'data' / 'input'
sys.path.append(str((BASE / 'src').resolve()))
from bq_utils import get_bigquery_client, fetch_tx_timestamps
```

```python
df = pd.read_csv(INPUT_DIR / 'seeds.csv')
txids = df['txid'].dropna().tolist()
len(txids)
```

```python
client = get_bigquery_client()
ts_map = fetch_tx_timestamps(client, txids)
df['block_timestamp'] = df['txid'].map(ts_map)
df.head()
```
