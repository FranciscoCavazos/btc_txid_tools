# Query seeds.csv using pandas-gbq

Runs a parameterised SQL query with pandas_gbq.read_gbq.

```python

from pathlib import Path
import pandas as pd
import pandas_gbq

base = Path('..').resolve()
seeds = pd.read_csv(base / 'data' / 'input' / 'seeds.csv')
txids = seeds['txid'].dropna().tolist()

sql = """
SELECT TO_HEX(`hash`) AS txid, block_timestamp
FROM `bigquery-public-data.crypto_bitcoin.transactions`
WHERE TO_HEX(`hash`) IN UNNEST(@txid_list)
"""
df = pandas_gbq.read_gbq(sql, project_id=None, location='US', params={'txid_list': txids})
df
```
